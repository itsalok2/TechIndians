```python
import pandas as pd 
import numpy as np 

def split_date(x):
    if pd.isna(x):
        return pd.Series([np.nan, np.nan, np.nan])
    
    x = str(x).split('.')[0]
    if len(x) == 8:
        return pd.Series([int(x[:4]),int(x[4:6]),int(x[6:])])
    elif len(x) == 6:
        return pd.Series([int(x[:4]),int(x[4:6]),np.nan])
    elif len(x) == 4:
        return pd.Series([int(x[:4]), np.nan, np.nan])
    else:
        return pd.Series([np.nan, np.nan, np.nan])
```
