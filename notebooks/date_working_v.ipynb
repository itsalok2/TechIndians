```python
import os
import pandas as pd
root_folder = 'faers_ascii21'
dataframe = {}
for folder in os.listdir(root_folder):
    path = os.path.join(root_folder,folder,'ASCII')
    for file in os.listdir(path):
        if file.endswith('.txt'):
            file_path = os.path.join(path,file)
            print(file_path,file.rsplit('.',1)[0])
            df_key = file.rsplit('.',1)[0].lower()
            dataframe[df_key] = pd.read_csv(file_path,sep='$',encoding = 'latin1',low_memory=False)
```
